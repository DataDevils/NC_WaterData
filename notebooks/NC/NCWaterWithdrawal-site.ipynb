## Extract data from a report
Example: http://www.ncwater.org/Permits_and_Registration/Water_Withdrawal_and_Transfer_Registration/report/view/0292-0001/2017
* Faciliy information
 * Name, Type, Sub-basin
* Withdrawal information
 * Avg daily withdrawal
* Discharge information
* Sub-basin transfer information


```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
siteID = '0292-0001'
year = 'year'
url = 'http://www.ncwater.org/Permits_and_Registration/Water_Withdrawal_and_Transfer_Registration/report/view/{0}/{1}'.format(siteID,year)
```

```python
#Read the tables in the HTML
dfs = pd.read_html(url,na_values='NaN')
```

```python
def unstackTable(df):
    #Copy the table
    df2 = df.copy(deep=True)
    
    #Convert the first row to columns, then drop the row
    colNames = df2.iloc[0]
    df2.columns = colNames
    df2.drop(0,inplace=True)

    #Convert two column format to one
    df2a = df2.iloc[:,:4]
    df2b = df2.iloc[:,4:]
    df2 = df2a.append(df2b)

    #Set month to be the index
    df2.set_index("Month",inplace=True)
    df2

    #Convert data types for columns 2, 3, and 4 (days, avg, max)
    df2.iloc[:,0] = df2.iloc[:,0].fillna(0).astype(int)
    df2.iloc[:,1] = df2.iloc[:,1].astype(float)
    df2.iloc[:,2] = df2.iloc[:,2].astype(float)
    
    #Return the table
    return df2
```

```python
dfFacility = dfs[2]

dfWithdrawal = unstackTable(dfs[3])
dfSource = dfs[4]

dfDischarge = unstackTable(dfs[5])
dfTransfer= unstackTable(dfs[8])
dfWater = pd.concat([dfWithdrawal,dfDischarge,dfTransfer], axis=1)
```
